```python
#y={1 x<10 0   else 
import matplotlib.pyplot as plt
x=[num for num in range(0,20)]
y=[]
for i in x:
    if i>=10:
        y.append(1)
    else:
        y.append(0)
plt.plot(x,y)
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.title("y=1 if x>=10 else y=0")
plt.show()


```

```python
#y=eax for different values of a.
import numpy as np
import matplotlib.pyplot as plt
num=[num for num in range(0,10)]
a=2 #constant value
x=np.array(num)*2
y=np.exp(x)
plt.plot(x,y)
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.title("y=e**ax")
plt.show()

```

```python
#y=7x2+3x+10 for 2≤x≤5
import matplotlib.pyplot as plt
x=[x for x in range(2,6)]
y=[]
for i in x:
    y.append(7*pow(i,2)+3*i+10)
plt.plot(x,y)
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.title('y=7x^2+3x+10')
plt.show()

```

```python
#y=11+e-x
import matplotlib.pyplot as plt
import numpy as np
x=[x for x in range(0,20)]
p=np.array(x)
p=1+np.exp(-p)
y=1/p
plt.plot(x,y)
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.show()

```

```python
#y=tan hx 
import numpy as np
import matplotlib.pyplot as plt
in_array = np.linspace(0, np.pi, 12)
h=2
out_array =h*np.tan(in_array)
print("in_array : ", in_array)
print("\nout_array : ", out_array)
# red for numpy.tan()
plt.plot(in_array, out_array, color='red', marker="o")
plt.title("numpy.tan()")
plt.xlabel("X")
plt.ylabel("Y")
plt.show()
```
