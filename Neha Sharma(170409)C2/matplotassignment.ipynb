# histogram: show distribution and used to plot frequency

# scatter plot: plot points 

```python
x=[1,1.5,2,2.5,3,3.5,3.6]
y=[7.5,8,8.5,9,9.5,10,10.5]

x1=[8,8.5,9,9.5,10,10.5,11]
y1=[3,3.5,3.7,4,4.5,5,5.2]

plt.scatter(x,y,label='high income low saving',color='r')
plt.scatter(x1,y1,label='low income high saving',color='b')
plt.xlabel('saving*100')
plt.ylabel('income*1000')
plt.title('Scatter Plot')
plt.legend()
plt.show()
```

# area plot

```python
days=[1,2,3,4,5]
sleeping=[6,7,8,9,10]
eating=[2,3,4,3,2]
working=[7,8,7,2,2]
playing=[8,5,7,8,13]

plt.plot([],[],color='m',label="sleeping",linewidth=5)  #[][]show 2d array
plt.plot([],[],color='c',label="eating",linewidth=5)
plt.plot([],[],color='r',label="working",linewidth=5)
plt.plot([],[],color='k',label="playing",linewidth=5)

plt.stackplot(days,sleeping,eating,working,playing,colors=['m','c','r','k'])

plt.xlabel('x')
plt.ylabel('y')
plt.title('stack plot')
plt.legend()
plt.show()
```

```python
days=[1,2,3,4,5]
sleeping=[6,7,8,9,10]
eating=[2,3,4,3,2]
working=[7,8,7,2,2]
playing=[8,5,7,8,13]
slices=[7,2,2,13]
activities=['sleeping','eating','working','playing']
cols=['c','m','r','b']

plt.pie(slices,labels=activities,colors=cols,startangle=90,shadow=True,explode=(0,0.1,0,0),autopct='%1.1f%%')

plt.title('Pie Plot')
plt.show()
```

# Multiple plots

```python
def f(t):
    return np.exp(-t) * np.cos(2*np.pi*t)
t1=np.arange(0.0,5.0,0.1)
t2=np.arange(0.0,5.0,0.02)
plt.subplot(211)  #first one show that there are 2 plots , second show that the plot are represented row wise or coloumn wise,third one show that 
#this is first/second plot
plt.plot(t1,f(t1),'bo',t2,f(t2))
plt.subplot(212)
plt.plot(t2,np.cos(2*np.pi*t2))
plt.show()
```
